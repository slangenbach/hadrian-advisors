## I. MNIST Classification using LSTM

```python
import os
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.datasets import mnist
from keras.utils import np_utils
```

```python
# Hyper parameters
batch_size = 128
nb_epoch = 10

# Parameters for MNIST dataset
img_rows, img_cols = 28, 28
nb_classes = 10

# Parameters for LSTM network
output_dim = 30
nb_time_steps = img_rows
dim_input_vector = img_cols
```

```python
# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('X_train original shape:', X_train.shape)
input_shape = (nb_time_steps, dim_input_vector)

X_train = X_train.astype('float32') / 255.
X_test = X_test.astype('float32') / 255.
Y_train = np_utils.to_categorical(y_train, nb_classes)
Y_test = np_utils.to_categorical(y_test, nb_classes)

print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# Build LSTM network
model = Sequential()
model.add(LSTM(output_dim, input_shape=input_shape))
model.add(Dense(nb_classes, activation='softmax', kernel_initializer='TruncatedNormal',
                bias_initializer='TruncatedNormal'))
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# Train
model.fit(X_train, Y_train, nb_epoch=nb_epoch, batch_size=batch_size, shuffle=True, verbose=1)
```

```python
# Evaluate
evaluation = model.evaluate(X_test, Y_test, batch_size=batch_size, verbose=1)
print('Summary: Loss over the test dataset: %.2f, Accuracy: %.2f' % (evaluation[0], evaluation[1]))
```

## II. Twitter Sentiment analysis using Word2Vec and LSTM

```python
import keras.backend as K
import multiprocessing
import tensorflow as tf
import numpy as np

from gensim.models.word2vec import Word2Vec

from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Flatten
from keras.layers import LSTM, Dense
from keras.optimizers import Adam

from nltk.stem.lancaster import LancasterStemmer
from nltk.tokenize import RegexpTokenizer
```

```python
mypath = os.path.abspath('./data/')
pathTwitter = os.path.join(mypath,'Twitter\\dataset.csv')
```

```python

corpus = []
labels = []
nRows = 50000

# Parse tweets and sentiments
with open(pathTwitter, 'r', encoding='utf-8') as df:
    for i, line in enumerate(df):
        if i >= nRows+1:
            break;
        if i == 0:
            # Skip the header
            continue

        parts = line.strip().split(',')
        
        # Sentiment (0 = Negative, 1 = Positive)
        labels.append(int(parts[1].strip()))
        
        # Tweet
        tweet = parts[3].strip()
        if tweet.startswith('"'):
            tweet = tweet[1:]
        if tweet.endswith('"'):
            tweet = tweet[::-1]
        
        corpus.append(tweet.strip().lower())
        
print('Corpus size: {}'.format(len(corpus)))
```

```python
# Tokenize and stem
tkr = RegexpTokenizer('[a-zA-Z0-9@]+')
stemmer = LancasterStemmer()

tokenized_corpus = []

for i, tweet in enumerate(corpus):
    tokens = [stemmer.stem(t) for t in tkr.tokenize(tweet) if not t.startswith('@')]
    tokenized_corpus.append(tokens)
```

```python
modelTwitter = Word2Vec.load(mypath + "modelTWITTER")
# Copy word vectors and delete Word2Vec model  and original corpus to save memory
X_vecs = modelTwitter.wv
```

```python
# Train subset size (0 < size < len(tokenized_corpus))
train_size = 40000

# Test subset size (0 < size < len(tokenized_corpus) - train_size)
test_size = 10000

# Compute average and max tweet length
avg_length = 0.0
max_length = 0

for tweet in tokenized_corpus:
    if len(tweet) > max_length:
        max_length = len(tweet)
    avg_length += float(len(tweet))
    
print('Average tweet length: {}'.format(avg_length / float(len(tokenized_corpus))))
print('Max tweet length: {}'.format(max_length))
```

```python
vector_size = modelTwitter.layer1_size
```

```python
# Tweet max length (number of tokens)
max_tweet_length = 15

# Create train and test sets
# Generate random indexes
indexes = set(np.random.choice(len(tokenized_corpus), train_size + test_size, replace=False))

X_train = np.zeros((train_size, max_tweet_length, vector_size), dtype=K.floatx())
Y_train = np.zeros((train_size, 2), dtype=np.int32)
X_test = np.zeros((test_size, max_tweet_length, vector_size), dtype=K.floatx())
Y_test = np.zeros((test_size, 2), dtype=np.int32)

for i, index in enumerate(indexes):
    for t, token in enumerate(tokenized_corpus[index]):
        if t >= max_tweet_length:
            break
        
        if token not in X_vecs:
            continue
    
        if i < train_size:
            X_train[i, t, :] = X_vecs[token]
        else:
            X_test[i - train_size, t, :] = X_vecs[token]
            
    if i < train_size:
        Y_train[i, :] = [1.0, 0.0] if labels[index] == 0 else [0.0, 1.0]
    else:
        Y_test[i - train_size, :] = [1.0, 0.0] if labels[index] == 0 else [0.0, 1.0]
```

```python
# Keras LSTM model
batch_size = 100
nb_epochs = 100

model = Sequential()

model.add(LSTM(64,input_shape=(max_tweet_length, vector_size)))
model.add(Dropout(0.2))
model.add(Dense(2, activation='softmax'))


# Compile the model
model.compile(loss='binary_crossentropy',
              optimizer=Adam(lr=0.0001, decay=1e-6),
              metrics=['accuracy'])
model.summary()
```

```python
batch_size = 50
nb_epochs = 100
# Fit the model
model.fit(X_train, Y_train,
          batch_size=batch_size,
          shuffle=True,
          epochs=nb_epochs,
          validation_data=(X_test, Y_test),
          callbacks=[EarlyStopping(min_delta=0.0025, patience=5)])
```

```python
# Evaluate
evaluation = model.evaluate(X_test, Y_test, batch_size=batch_size, verbose=0)
print('Summary: Accuracy over the test dataset: Accuracy: %.2f' % evaluation[1])
```

```python
import seaborn as sns
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from pandas import datetime
import time
import itertools
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
from math import sqrt
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.layers.recurrent import LSTM
from keras.models import load_model
import keras
from keras import backend as K
```

```python
def PreprocessData(Tab,ma):
    Tab['Yield'].fillna(method='ffill',inplace=True)
    Tab = Tab.dropna(subset=['Yield'],axis=0)
    
    # Percentage change
    Tab['Pct'] = Tab['Yield'].pct_change()
    
    # Moving Average    
    if ma != []:
        for moving in ma:
            Tab['{}ma'.format(moving)] = Tab['Yield'].rolling(window=moving).mean()
    
    Tab.fillna(method='ffill',inplace=True)
    Tab = Tab.dropna(axis=0)
    
    ## Normalize data
    min_max_scaler = preprocessing.MinMaxScaler()
    Tab['Yield'] = min_max_scaler.fit_transform(Tab.Yield.values.reshape(-1,1))
    Tab['Date'] = Tab['Date'].apply(lambda elt: datetime.strptime(elt,'%Y-%m-%d').date())
    
    cols = Tab.columns.tolist()
    cols.insert(len(cols)-1, cols.pop(cols.index('Yield')))
    Tab = Tab[cols]
    return Tab
```

**We want to use 12 rows of the dataset to predict the 13th normalized yield, so we are going to create a 3d matrix (N,12,number_features) to predict a vector of size (N)**

Each 12 x number_features of days t-12,...,t will be used to predict the yield at time t+1.

```python
height = 12
number_features = len(Tab.columns)-1
shape = [height, number_features, 1]

## Ouput dimensions (number of neurons)
output_dim1 = 256
output_dim2 = 256
output_dim3 = 32
output_dim4 = 1

dropout = 0.3
decay = 0.5
epochs = 5
stock_name = 'EURUSD'
```

```python
def Split_Train_Test(Tab,height,number_features,propLearning):
    
    arrayData = Tab.values[:,1:].copy()
    newheight = height + 1 # index starting from 0
    result = []

    for index in range(len(arrayData) - newheight): # maxmimum date = lastest date - height
        result.append(arrayData[index: index + newheight]) # index : index + height

    result = np.array(result)
    row = round(propLearning * result.shape[0]) # propLearning to split

    train = result[:int(row), :]
    X_train = train[:, :-1] # all data until day n
    y_train = train[:, -1][:,-1] # day n + 1 normalized yield

    X_test = result[int(row):, :-1]
    y_test = result[int(row):, -1][:,-1]

    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], number_features))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], number_features))
    
    return X_train, X_test, y_train, y_test
```

```python
propLearning = 0.8
X_train, X_test, y_train, y_test = Split_Train_Test(Tab,height,number_features,propLearning)
```

```python
def PnLGivenOurPredictions(model, Tab, X_test, y_test,
                           column_name, horizon=1):
    '''
    Computes the Profit and Loss values given a model, a table of prices and a testing set
    here it is for a regression model (pricing)
    '''
    Yield = Tab[column_name].copy()
    Date = Tab['Date'].copy()
    Date = Date.apply(lambda elt: datetime.strptime(elt,'%Y-%m-%d').date())
    
    nb_rows_test = len(y_test)
    Ret = Yield.pct_change(horizon)
    
    p = model.predict(X_test)
    p = p.T[0]
    
    p_ts = pd.Series(p)
    p_ts = p_ts.pct_change(horizon)
    PnL = (1 + np.sign(p_ts).values * Ret[-nb_rows_test:]).cumprod()
    PnL.name = 'Yield'
    PnL.index = Date.values[-nb_rows_test:]
    return PnL
    
```

### Let's try an other method to process the data and to train, avoiding future information.

```python
def PreprocessDataWithoutFutureInfo(Tab_PrevDays,Tab,ma):
    Tab_PrevDays['Yield'].fillna(method='ffill',inplace=True)
    Tab_PrevDays.dropna(subset=['Yield'],axis=0,inplace=True)
    Tab['Yield'].fillna(method='ffill',inplace=True)
    
    # Percentage change
    Tab_PrevDays['Pct'] = Tab_PrevDays['Yield'].pct_change()
    Tab['Pct'] = Tab['Yield'].pct_change()
    
    # Moving Average    
    if ma != []:
        for moving in ma:
            Tab['{}ma'.format(moving)] = Tab['Yield'].rolling(window=moving).mean()
            Tab_PrevDays['{}ma'.format(moving)] = Tab_PrevDays['Yield'].rolling(window=moving).mean()
            
    # AutoCorrelation
    autocorr=[5,10,50,100,250,500]
    for auto in autocorr:
        Tab['{}autocorr'.format(auto)] = Tab['Yield'].autocorr(lag=auto)
        Tab_PrevDays['{}autocorr'.format(auto)] = Tab_PrevDays['Yield'].autocorr(lag=auto)
    
    # Exponential Weighted Moving average
    spans = [2,5,7,20]
    for sp in spans:
        Tab['{}ewma'.format(sp)] = Tab['Yield'].ewm(span=sp).mean()
        Tab['{}ewstd'.format(sp)] = Tab['Yield'].ewm(span=sp).std()
        Tab_PrevDays['{}ewma'.format(sp)] = Tab_PrevDays['Yield'].ewm(span=sp).mean()
        Tab_PrevDays['{}ewstd'.format(sp)] = Tab_PrevDays['Yield'].ewm(span=sp).std()
    
    Tab = Tab.dropna(axis=0)
    Tab_PrevDays.fillna(0.0,inplace=True)
    
    ## Quantilize pct yield to get discrete label
    qTile70 = Tab_PrevDays['Pct'].quantile(.7)
    qTile30 = Tab_PrevDays['Pct'].quantile(.3)
    Tab['Output'] = 1 * (Tab['Pct'] > qTile70) - 1 * (Tab['Pct'] <= qTile30)
    
    ## Normalize output data using Tab with Previous data
    min_max_scaler = preprocessing.MinMaxScaler()
    min_max_scaler.fit(Tab_PrevDays['Yield'].values.reshape(-1,1));
    
    ## Normalize data using Tab with Previous data
    min_max_scaler_features = preprocessing.MinMaxScaler()
    min_max_scaler_features.fit(Tab_PrevDays.values[:,2:]);
    
    ## Transform Tab
    Tab['Yield'] = min_max_scaler.transform(Tab['Yield'].values.reshape(-1,1))
    Tab['Date'] = Tab['Date'].apply(lambda elt: datetime.strptime(elt,'%Y-%m-%d').date())
    
    listCols2Standardize = [col for col in Tab.columns if 'Date' not in col \
        and 'Output' not in col and 'Yield' not in col]
    Tab[listCols2Standardize] = min_max_scaler_features.transform(Tab[listCols2Standardize].values)
    
    cols = Tab.columns.tolist()
    cols.insert(len(cols)-1, cols.pop(cols.index('Output')))
    Tab = Tab[cols]
    return Tab
```

```python
def PnLFromClassifications(model, Tab, X_test, y_test,
                           column_name, horizon=1):
    '''
    Computes the Profit and Loss given a model, a table of prices and a testing set
    here it is for a classification model (predicting the directions of the days, long or short)
    '''
    Yield = Tab[column_name].copy()
    Date = Tab['Date'].copy()
    Date = Date.apply(lambda elt: datetime.strptime(elt,'%Y-%m-%d').date())
    
    nb_rows_test = len(y_test)
    Ret = Yield.pct_change(horizon)
    
    p = model.predict(X_test)
    p = np.argmax(p,axis=1) - 1
    
    PnL = (1 + p * Ret[-nb_rows_test:]).cumprod()
    PnL.name = 'Yield'
    PnL.index = Date.values[-nb_rows_test:]
    return PnL
```
